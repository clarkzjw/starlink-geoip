```python
from pathlib import Path
from pprint import pprint
import os
import json


latency = {}

def load_residential(filename: Path):
    data = json.loads(filename.read_text())
    country_data = data["admin0Metrics"]
    return country_data


path = Path("/home/clarkzjw/starlink-map/latency/metrics_residential/")
for dirpath, dirnames, files in os.walk(path):
    if len(files) != 0:
        for f in files:
            date = dirpath.split("/")[-1]
            data = load_residential(Path(dirpath) / f)
            latency[date] = data

latency = dict(sorted(latency.items()))

unique_country_code = []
for k, v in latency.items():
    unique_country_code += [x for x in v]

unique_country_code = sorted(list(set(unique_country_code)))

result = dict.fromkeys(unique_country_code)

for country_code in unique_country_code:
    for date, v in latency.items():
        if country_code not in v:
            continue
        else:
            if result[country_code] is None:
                result[country_code] = {}
            result[country_code][date] = v[country_code]


result = dict(sorted(result.items()))
with open("./latency/residential.json", "w") as f:
    json.dump(result, f, indent=4) 
```

```python
from matplotlib import pyplot as plt
from pprint import pprint
import json


with open("./latency/residential.json", "r") as f:
    data = json.load(f)
    
    for key in data.keys():
        print(key)
        result = data[key]
        date = [x[4:8] for x in result.keys()]
        
        fig = plt.figure(figsize=(10, 15))
        # latency
        latency_p20 = [x["latency_p20"] for x in result.values()]
        latency_p50 = [x["latency_p50"] for x in result.values()]
        latency_p80 = [x["latency_p80"] for x in result.values()]
        
        ax1 = fig.add_subplot(211)
        ax1.plot(date, latency_p20, label="latency_p20")
        ax1.plot(date, latency_p50, label="latency_p50")
        ax1.plot(date, latency_p80, label="latency_p80")
        ax1.legend()
        plt.xticks(rotation=45)
        ax1.set_title(key)
        ax1.set_xlabel("Date")
        ax1.set_ylabel("Latency (ms)")
        
        # download
        download_p20 = [x["download_p20"] for x in result.values()]
        download_p50 = [x["download_p50"] for x in result.values()]
        download_p80 = [x["download_p80"] for x in result.values()]
        
        ax2 = fig.add_subplot(212)
        ax2.plot(date, download_p20, label="download_p20")
        ax2.plot(date, download_p50, label="download_p50")
        ax2.plot(date, download_p80, label="download_p80")
        
        # upload
        upload_p20 = [x["upload_p20"] for x in result.values()]
        upload_p50 = [x["upload_p50"] for x in result.values()]
        upload_p80 = [x["upload_p80"] for x in result.values()]
        
        ax2.plot(date, upload_p20, label="upload_p20")
        ax2.plot(date, upload_p50, label="upload_p50")
        ax2.plot(date, upload_p80, label="upload_p80")
        ax2.legend()
        plt.xticks(rotation=45)
        ax2.set_title(key)
        ax2.set_xlabel("Date")
        ax2.set_ylabel("Mbps")
        
        # plt.show()
        plt.savefig(f"./latency/metrics_residential/figures/{key}.png")
        plt.close()
    
```
